# 🧙‍♂️ **마법사 모험: 튜플 훈련**
### 당신은 마법사 학교를 졸업하고 마법사 탐험대에 합류했습니다. 다양한 미션을 수행하며 **자료형**을 다루는 능력을 연습해야 합니다.
###문제를 풀며 모험을 완수하세요!

# 🪄 튜플 훈련

## [문제 1] 변경 불가능한 데이터 생성
###  "빗자루", "망토", "수정구슬" 값이 들어간 튜플을 생성하세요.
### 사용할 변수명 : `in_my_bag`

## [문제 2] 튜플 값 확인
###  `in_my_bag` 튜플에 "망토"가 있는지 확인하세요.

## [문제 3] 튜플 길이 확인
### `in_my_bag` 튜플의 길이를 출력하세요.

## [문제 4] 인덱싱
### `in_my_bag` 튜플에서 "빗자루"의 인덱스 번호를 출력하세요.

## [문제 5] 튜플 슬라이싱
### ("빗자루", "망토", "수정구슬", "지팡이",)에서 "망토", "수정구슬"만 추출하세요.

## [문제 6] 튜플 값 변경 시도
### ("빗자루", "망토",)에서 "망토"를 "지팡이"로 바꿀 수 있는지 확인하세요.

## [문제 7] 튜플 병합
### ("빗자루", "망토",)와 ("수정구슬",)을 결합하세요.

## [문제 8] 튜플에서 값 찾기
### `in_my_bag` 튜플에서 "지팡이"의 위치를 출력하세요

## [문제 9] 튜플에서 마지막 항목 확인
### `in_my_bag` 튜플에서 마지막 값을 출력하세요.

## [문제 10] 튜플 길이 비교
### `in_my_bag` 튜플과 ("빗자루", "망토", "수정구슬", "마법 책", "지팡이") 중 더 긴 튜플을 찾으세요.

## [문제 11] 패킹
### 1, 2, "오즈", "패킹" 을 하나의 변수 `p`에 담고, p의 자료형이 무엇인지 출력해주세요.

## [문제 12] 인덱스 확인
### `p` 에서 "오즈" 의 인덱스를 출력하세요.

## [문제 13] 튜플과 문자열 비교
("초코송이","오징어집","초코파이","쿠크다스") 의 자료형 타입과

("초코송이","오징어집","초코파이","쿠크다스",) 자료형 타입을 비교하고 차이가 있다면 그 이유를 주석으로 서술하세요.
