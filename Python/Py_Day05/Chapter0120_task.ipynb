__[과제 3: 주문 처리 메서드 process_order]__

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
# __BungeoppangShop 클래스__

# - __init__(): 초기 재고, 가격, 판매량을 설정합니다.
# - check_stock(): 현재 재고를 출력합니다.
# - process_order(): 주문을 처리하고 재고를 업데이트합니다.
# - admin_mode(): 관리자 모드에서 재고를 추가합니다.
# - calculate_total_sales(): 총 매출을 계산하고 출력합니다.

class BungeoppangShop:
    def __init__(self): # 현재재고
        self.stock = {"팥붕어빵":10, "슈크림붕어빵":8, "초코붕어빵":5}
        self.prices = {"팥붕어빵":1000, "슈크림붕어빵":1200, "초코붕어빵":1500}
        self.sales = {"팥붕어빵":0, "슈크림붕어빵":0, "초코붕어빵":0}

    def check_stock(self):  # 현재재고 확인용
        print(f'[재고] 현재 재고는 {self.stock}')

    def process_order(self, bread_type, bread_count): # 주문처리
        print(f'[주문] 고객님의 주문은 {bread_type} {bread_count}개 입니다.') # 주문을 확인한다.

        try: # 붕어빵종류 예외처리
            if self.stock.get(bread_type) >= bread_count: # 현재재고와 비교한다. (주문수량보다 재고가 많을 경우 // 주문 진행)                
                self.stock.update({bread_type : self.stock.get(bread_type) - bread_count}) # 현재재고업데이트
                self.stock.update({bread_type : self.sales.get(bread_type) + bread_count}) # 총주문수량업데이트   
                print(f'[주문] 주문해주셔서 감사합니다!')
                print(f'[재고] 현재 주문 가능 수량은 {self.stock} 입니다!')
            else: # 현재재고와 비교한다. (주문수량보다 재고가 적을 경우 // 주문 미이행)
                print("[주문] 재고수량이 부족합니다ㅠㅠ 다시 주문 해 주세요.")
                print(f'[재고] 현재 주문 가능 수량은 {self.stock} 입니다!')
        except TypeError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생
            print("주문을 다시 확인해주세요!")      
        except NameError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생            
            print("주문을 다시 확인해주세요!")  

work = BungeoppangShop()
work.check_stock()
work.process_order("팥붕어빵", 5)

```

```python
# __BungeoppangShop 클래스__

# - __init__(): 초기 재고, 가격, 판매량을 설정합니다.
# - check_stock(): 현재 재고를 출력합니다.
# - process_order(): 주문을 처리하고 재고를 업데이트합니다.
# - admin_mode(): 관리자 모드에서 재고를 추가합니다.
# - calculate_total_sales(): 총 매출을 계산하고 출력합니다.

class BungeoppangShop:
    def __init__(self): # 현재재고
        self.stock = {"팥붕어빵":10, "슈크림붕어빵":8, "초코붕어빵":5}
        self.prices = {"팥붕어빵":1000, "슈크림붕어빵":1200, "초코붕어빵":1500}
        self.sales = {"팥붕어빵":0, "슈크림붕어빵":0, "초코붕어빵":0}

    def check_stock(self):  # 현재재고 확인용
        print('---------- 관리자모드 ----------')
        print(f'[재고/확인] 현재 재고는 {self.stock}')

    def process_order(self, bread_type, bread_count): # 주문처리
        print('---------- 주문 ----------')
        print(f'[주문/접수] 고객님의 주문은 {bread_type} {bread_count}개 입니다.') # 주문을 확인한다.

        try: # 입력값 예외처리
            if self.stock.get(bread_type) >= bread_count: # 현재재고와 비교한다. (주문수량보다 재고가 많을 경우 // 주문 진행)                
                self.stock.update({bread_type : self.stock.get(bread_type) - bread_count}) # 현재재고업데이트 (현재재고 - 주문수량)
                self.sales.update({bread_type : self.sales.get(bread_type) + bread_count}) # 총주문수량업데이트 (총주문수량 + 현재주문수량)  
                print(f'[주문/완료] 주문해주셔서 감사합니다!')
                print(f'[재고/확인] 현재 주문 가능 수량은 {self.stock} 입니다!')
            else: # 현재재고와 비교한다. (주문수량보다 재고가 적을 경우 // 주문 미이행)
                print("[주문/재고부족] 재고수량이 부족합니다ㅠㅠ 다시 주문 해 주세요.")
                print(f'[재고/확인] 현재 주문 가능 수량은 {self.stock} 입니다!')            
        except TypeError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생
            print("[주문/오류] 주문을 다시 확인해주세요!")      
        except NameError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생            
            print("[주문/오류] 주문을 다시 확인해주세요!")  

    def admin_mode(self, bread_type, additional_stock): # 관리자모드 (재고관리)
        print('---------- 관리자모드 ----------')
        print(f'[재고/추가] 추가된 재고는 {bread_type} {additional_stock}개 입니다.') # 추가된 재고를 확인한다.

        try: # 입력값 예외처리
            self.stock.update({bread_type : self.stock.get(bread_type) + additional_stock}) # 현재재고업데이트 (현재재고 + 추가된재고)
            print(f'[재고/확인] 현재 주문 가능 수량은 {self.stock} 입니다!')
        except TypeError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생
            print("[재고/오류] 주문을 다시 확인해주세요!")      
        except NameError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생            
            print("[재고/오류] 주문을 다시 확인해주세요!")  

    def calculate_total_sales(self):    # 관리자모드 (총매출확인)
        print('---------- 관리자모드 ----------')
        print(f'[매출/확인] 오늘의 총 주문 수량은 {self.sales} 입니다!')  
        total_sales = dict()    # 총 매출을 담을 딕셔너리 선언

        for bread_type in self.prices.keys(): # 붕어빵 종류 할당
            total_sales.update({bread_type : self.prices.get(bread_type) * self.sales.get(bread_type)}) # 붕어빵 종류별로 매출 업데이트
        
        total_pay = list(total_sales.values())[0] + list(total_sales.values())[1] + list(total_sales.values())[2]
        print(f'[매출/확인] 오늘의 총 매출은 {total_pay} 입니다!')



work = BungeoppangShop()
work.check_stock()
work.process_order("팥붕어빵", 3)
work.process_order("초코붕어빵", 3)
work.admin_mode("초코붕어빵", 10)
work.calculate_total_sales()
```

- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
# __BungeoppangShop 클래스__

# - __init__(): 초기 재고, 가격, 판매량을 설정합니다.
# - check_stock(): 현재 재고를 출력합니다.
# - process_order(): 주문을 처리하고 재고를 업데이트합니다.
# - admin_mode(): 관리자 모드에서 재고를 추가합니다.
# - calculate_total_sales(): 총 매출을 계산하고 출력합니다.

from datetime import datetime # 현재시각 표시

class BungeoppangShop:
    def __init__(self): # 현재재고
        self.stock = {"팥붕어빵":10, "슈크림붕어빵":8, "초코붕어빵":5}
        self.prices = {"팥붕어빵":1000, "슈크림붕어빵":1200, "초코붕어빵":1500}
        self.sales = {"팥붕어빵":0, "슈크림붕어빵":0, "초코붕어빵":0}
        print('[오늘 장사시작]]')

    def check_stock(self):  # 현재재고 확인용
        print('---------- 관리자모드 ----------')
        print(f'[재고/확인] 현재 재고는 {self.stock}')

    def process_order(self, bread_type, bread_count, order_count): # 주문처리
        print('---------- 주문' + str(order_count) + ' ----------')
        print(f'[주문/접수] 주문번호 {order_count}번 고객님의 주문은 {bread_type} {bread_count}개 입니다.') # 주문을 확인한다.

        try: # 입력값 예외처리
            if self.stock.get(bread_type) >= bread_count: # 현재재고와 비교한다. (주문수량보다 재고가 많을 경우 // 주문 진행)                
                self.stock.update({bread_type : self.stock.get(bread_type) - bread_count}) # 현재재고업데이트 (현재재고 - 주문수량)
                self.sales.update({bread_type : self.sales.get(bread_type) + bread_count}) # 총주문수량업데이트 (총주문수량 + 현재주문수량)  
                print(f'[주문/완료] 주문해주셔서 감사합니다!')
                print(f'[재고/확인] 현재 주문 가능 수량은 {self.stock} 입니다!')
            else: # 현재재고와 비교한다. (주문수량보다 재고가 적을 경우 // 주문 미이행)
                print("[주문/재고부족] 재고수량이 부족합니다ㅠㅠ 다시 주문 해 주세요.")
                print(f'[재고/확인] 현재 주문 가능 수량은 {self.stock} 입니다!')
        except TypeError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생
            print("[주문/오류] 주문을 다시 확인해주세요!1")      
        except NameError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생            
            print("[주문/오류] 주문을 다시 확인해주세요!2")  

    def admin_mode(self, bread_type, additional_stock): # 관리자모드 (재고관리)
        print('---------- 관리자모드 ----------')
        print(f'[재고/추가] 추가된 재고는 {bread_type} {additional_stock}개 입니다.') # 추가된 재고를 확인한다.

        try: # 입력값 예외처리
            self.stock.update({bread_type : self.stock.get(bread_type) + additional_stock}) # 현재재고업데이트 (현재재고 + 추가된재고)
            print(f'[재고/확인] 현재 주문 가능 수량은 {self.stock} 입니다!')
        except TypeError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생
            print("[재고/오류] 주문을 다시 확인해주세요!")      
        except NameError: # 붕어빵이름 및 수량 잘못입력 시 예외처리 발생            
            print("[재고/오류] 주문을 다시 확인해주세요!")  

    def calculate_total_sales(self):    # 관리자모드 (총매출확인)
        print('---------- 관리자모드 ----------')
        print(f'[매출/확인] 오늘의 총 주문 수량은 {self.sales} 입니다!')  
        total_sales = dict()    # 총 매출을 담을 딕셔너리 선언
        
        for bread_type in self.prices.keys(): # 붕어빵 종류 할당
            total_sales.update({bread_type : self.prices.get(bread_type) * self.sales.get(bread_type)}) # 붕어빵 종류별로 매출 업데이트
        
        total_pay = list(total_sales.values())[0] + list(total_sales.values())[1] + list(total_sales.values())[2]
        print(f'[매출/확인] 오늘의 총 매출은 {total_pay}원입니다!')
        
        with open('sales.cvs', 'a') as f:
            f.write("[" + str(datetime.now()) + "] 오늘의 매출 : " + format(total_pay, ",") + " 원\n")
        print("[오늘 붕어빵 장사 끄읕~]")


def main():
    working = BungeoppangShop() # BungeoppangShop 클래스 생성
    order_count = 0 # 주문번호
    working.check_stock() # 현재재고 확인

    while True:
        doing = input("작업을 선택해 주세요! 주문(1) 관리자모드(2) 종료(3)") # 작업 입력 받음

        try:
            if doing == "주문" or doing == "1":
                order_count += 1
                bungeoppang_type, bungeoppang_amount = input("주문하실 붕어빵 종류와 수량을 입력해주세요! (팥붕어빵, 슈크림붕어빵, 초코붕어빵) (쉼표로 구분!!)").split(",")
                working.process_order(bungeoppang_type, int(bungeoppang_amount), order_count)
            elif doing == "관리자모드" or doing == "2":
                add_type, add_amount = input("재고를 추가할 붕어빵 종류와 수량을 입력해주세요! (팥붕어빵, 슈크림붕어빵, 초코붕어빵) (쉼표로 구분!!)").split(",")
                working.admin_mode(add_type, int(add_amount))
            elif doing == "종료" or doing == "3":
                working.calculate_total_sales()
                break       
        except TypeError: # 모드 잘못입력 시 예외처리 발생
            print("[모드/오류] 주문을 다시 확인해주세요!11")      
        except NameError: # 모드 잘못입력 시 예외처리 발생            
            print("[모드/오류] 주문을 다시 확인해주세요!22")  


if __name__ == "__main__":  # main() 실행
    main()
```
